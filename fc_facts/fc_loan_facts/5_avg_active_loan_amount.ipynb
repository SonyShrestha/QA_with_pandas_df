```python
# df_loan_facts
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_facts")
df_loan_facts=pd.read_sql('fc_loan_facts',con=engine)
df_loan_facts=df_loan_facts[["account_number","customer_code","avg_active_loan_amount"]]
```

```python
df_active_loan_customers=df_loan_master[df_loan_master["is_active"]==1][["customer_code","total_loan_amount"]].groupby(["customer_code"]).agg(
    avg_active_loan_amount=("total_loan_amount","mean")
)
df_active_loan_customers.reset_index(inplace=True)
df_active_loan_customers
```

```python
df_qa_vs_dev=pd.merge(df_active_loan_customers,df_loan_facts,how="inner",on="customer_code",suffixes=('_qa','_dev'))
df_qa_vs_dev["status"]=np.select(
    [
        df_qa_vs_dev["avg_active_loan_amount_qa"]==df_qa_vs_dev["avg_active_loan_amount_dev"]
    ],
    [
        'Pass'
    ],
    default='Fail'
)
df_qa_vs_dev
```
