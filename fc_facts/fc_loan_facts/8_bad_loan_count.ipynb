# Validation of bad_loan_count

```python
import sqlalchemy 
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.relativedelta import *
```

```python
end_date=datetime(2021,6,30)
start_date=end_date-relativedelta(months=11)
start_date=start_date.strftime('%Y-%m-01')
start_date
provision_type=[2,3,4,5]
```

```python
pd.set_option('display.float_format', '{:.10f}'.format)
```

```python
# df_loan_master
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_loan_master=pd.read_sql('fc_loan_facts',con=engine)
df_loan_master["is_bad_loan"]=np.select(
    [
        df_loan_master["provision_type"].isin(provision_type)
    ],
    [
        1
    ],
    default=0
)
df_loan_master
```

```python
# df_loan_facts
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_facts")
df_loan_facts=pd.read_sql('fc_loan_facts',con=engine)
df_loan_facts=df_loan_facts[["account_number","customer_code","bad_loan_count"]]
df_loan_facts
```

```python
df_loan_customers=df_loan_master[df_loan_master["is_bad_loan"]==1][["customer_code"]].groupby(["customer_code"]).agg(
    bad_loan_count=("customer_code","count")
)
df_loan_customers.reset_index(inplace=True)
df_loan_customers
```

```python
df_qa_vs_dev=pd.merge(df_loan_customers,df_loan_facts,how="inner",on="customer_code",suffixes=('_qa','_dev'))
df_qa_vs_dev["status"]=np.select(
    [
        df_qa_vs_dev["bad_loan_count_qa"]==df_qa_vs_dev["bad_loan_count_dev"]
    ],
    [
        'Pass'
    ],
    default='Fail'
)
df_qa_vs_dev
```
