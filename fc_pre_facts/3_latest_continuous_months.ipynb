```python
# Get selected fields
df_transaction_base=df_transaction_base[["account_number","tran_date","lcy_amount","is_salary"]] 
df_transaction_base_sal=df_transaction_base[df_transaction_base["is_salary"]!=0]
```

```python
# Read fc_pre_facts table
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_facts")
df_pre_facts=pd.read_sql('fc_pre_facts',con=engine)
```

```python
# Get selected fields
df_pre_facts=df_pre_facts[["account_number","latest_continuous_months","recency"]]
df_pre_facts
# df_pre_facts=df_pre_facts[df_pre_facts["account_number"]=="fffef9f15af92a8025906e7bd8550ba5d33666da"]
```

```python
# Date Format
df_transaction_base_sal["tran_date_1"]=df_transaction_base_sal["tran_date"].dt.strftime("%Y-%m-01")
df_transaction_base_sal
```

```python
# Get distinct account_number and tran_date
df_transaction_base_sal_dist=df_transaction_base_sal.drop_duplicates(subset=["account_number","tran_date_1"],keep="first")
df_transaction_base_sal_dist=df_transaction_base_sal_dist[["account_number","tran_date_1"]]
df_transaction_base_sal_dist.rename(columns={"tran_date_1":"tran_date"},inplace=True)
df_transaction_base_sal_dist.reset_index(inplace=True)
df_transaction_base_sal_dist.sort_values(by=["account_number","tran_date"],ascending=True,inplace=True)
df_transaction_base_sal_dist.reset_index(inplace=True)
df_transaction_base_sal_dist["end_date"]=end_date
df_transaction_base_sal_dist
# df_transaction_base_sal_dist=df_transaction_base_sal_dist[df_transaction_base_sal_dist["account_number"]=="fffef9f15af92a8025906e7bd8550ba5d33666da"]
```

```python
df_transaction_base_sal_dist["lag_tran_date"]=df_transaction_base_sal_dist.groupby(["account_number"])["tran_date"].shift(-1)
df_transaction_base_sal_dist["tran_date"]=pd.to_datetime(df_transaction_base_sal_dist["tran_date"])
df_transaction_base_sal_dist["lag_tran_date"]=pd.to_datetime(df_transaction_base_sal_dist["lag_tran_date"])
df_transaction_base_sal_dist["lag_tran_date"].dt.year
df_transaction_base_sal_dist["month_diff"]=((df_transaction_base_sal_dist["lag_tran_date"].dt.year-df_transaction_base_sal_dist["tran_date"].dt.year)*12+(df_transaction_base_sal_dist["lag_tran_date"].dt.month-df_transaction_base_sal_dist["tran_date"].dt.month)).fillna(1)
df_transaction_base_sal_dist=df_transaction_base_sal_dist[["account_number","tran_date","lag_tran_date","month_diff"]]
df_transaction_base_sal_dist["var"]=df_transaction_base_sal_dist["month_diff"].ne(df_transaction_base_sal_dist["month_diff"].shift()).cumsum()
df_transaction_base_sal_dist
```

```python
df_continuity_from_recency=df_transaction_base_sal_dist.groupby(["account_number","var"]).agg(
    continuity_from_recency_=("var","count")
)

df_continuity_from_recency.reset_index(inplace=True)
df_continuity_from_recency=df_continuity_from_recency.groupby(["account_number"]).agg(
    continuity_from_recency=("continuity_from_recency_","last")
)
df_continuity_from_recency
```

```python
df_qa_vs_dev=pd.merge(df_continuity_from_recency,df_pre_facts,on="account_number",suffixes=('_qa','_dev'),how="inner")
df_qa_vs_dev["latest_continuous_months_qa"]=np.select(
    [
        df_qa_vs_dev["recency"]==1
    ],
    [
        df_qa_vs_dev["continuity_from_recency"]
    ],
    default=0
)
df_qa_vs_dev.rename(columns={"latest_continuous_months":"latest_continuous_months_dev"},inplace=True)
df_qa_vs_dev["latest_continuous_months_status"]=np.select(
    [
        df_qa_vs_dev["latest_continuous_months_dev"]==df_qa_vs_dev["latest_continuous_months_qa"]
    ],
    [
        "pass"
    ],
    default="fail"
)
df_qa_vs_dev
```
