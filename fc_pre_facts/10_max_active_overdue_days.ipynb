```python
import sqlalchemy
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# Read loan_master
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_loan_master=pd.read_sql('fc_loan_facts',con=engine)
df_loan_master
```

```python
# Read fc_pre_facts
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_facts")
df_pre_facts=pd.read_sql('fc_pre_facts',con=engine)
df_pre_facts=df_pre_facts[["account_number","customer_code","max_active_overdue_days"]]
df_pre_facts
```

```python
df_loan_master_selected=df_loan_master[df_loan_master["is_active"]==1]
df_loan_master_selected=df_loan_master_selected[["customer_code","overdue_days"]]
df_loan_master_selected=df_loan_master_selected.groupby(["customer_code"]).agg(
    max_active_overdue_days=("overdue_days","max")
)
df_loan_master_selected.reset_index(inplace=True)
df_loan_master_selected=df_loan_master_selected[df_loan_master_selected["max_active_overdue_days"]!=0]
df_loan_master_selected
```

```python
df_qa_vs_dev=pd.merge(df_loan_master_selected,df_pre_facts,how="inner",on="customer_code",suffixes=('_qa','_dev'))
df_qa_vs_dev["status"]=np.select(
    [
        df_qa_vs_dev["max_active_overdue_days_qa"]==df_qa_vs_dev["max_active_overdue_days_dev"]
    ],
    [
        "pass"
    ],
    default="fail"
)
df_qa_vs_dev
```
