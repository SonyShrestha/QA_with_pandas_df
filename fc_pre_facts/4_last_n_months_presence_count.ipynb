```python
import sqlalchemy
import pandas as pd
import numpy as np
from datetime import datetime
import dateutil.relativedelta
```

```python
end_date=datetime(2021,7,1)
run_date=datetime(2021,9,9)
start_date=end_date-dateutil.relativedelta.relativedelta(months=5)
start_date
```

```python
# Read fc_account_master table
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_account_master=pd.read_sql('fc_account_master',con=engine)
```

```python
# Get selected fields
df_account_master=df_account_master[["account_number","customer_code","acc_open_date","acc_closed_date","active_flag"]]
df_account_master
```

```python
# Read fc_transaction_base table
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_transaction_base=pd.read_sql('fc_transaction_base',con=engine)
```

```python
# Get selected fields
df_transaction_base=df_transaction_base[["account_number","tran_date","lcy_amount","is_salary"]] 
df_transaction_base_sal=df_transaction_base[df_transaction_base["is_salary"]!=0]
```

```python
# Read fc_pre_facts table
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_facts")
df_pre_facts=pd.read_sql('fc_pre_facts',con=engine)
```

```python
# Get selected fields
df_pre_facts=df_pre_facts[["account_number","last_n_months_presence_count"]]
df_pre_facts
```

```python
df_transaction_base_sal_selected_date=df_transaction_base_sal[(df_transaction_base_sal["tran_date"]>=start_date)&(df_transaction_base_sal["tran_date"]<=end_date)]
df_transaction_base_sal_selected_date["year"]=df_transaction_base_sal_selected_date["tran_date"].dt.year
df_transaction_base_sal_selected_date["month"]=df_transaction_base_sal_selected_date["tran_date"].dt.month
df_transaction_base_sal_selected_date["year_month"]=df_transaction_base_sal_selected_date["year"].astype(str)+df_transaction_base_sal_selected_date["month"].astype(str)
df_transaction_base_sal_selected_date
```

```python
df_last_n_months_presence_count=df_transaction_base_sal_selected_date.groupby(["account_number"]).agg(
        last_n_months_presence_count=("year_month","nunique")
)
df_last_n_months_presence_count
```

```python
df_qa_vs_dev=pd.merge(df_last_n_months_presence_count,df_pre_facts,how="inner",on="account_number",suffixes=('_qa','_dev'))
df_qa_vs_dev["status"]=np.select(
    [
        df_qa_vs_dev["last_n_months_presence_count_qa"]==df_qa_vs_dev["last_n_months_presence_count_dev"]
    ],
    [
        "pass"
    ],
    default="fail"
)
df_qa_vs_dev
```
