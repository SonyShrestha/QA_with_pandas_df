# Validation of in_lien_fd

```python
import sqlalchemy
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
run_date=datetime(2021,9,10)
```

```python
# Read fd_master
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_fd_master=pd.read_sql('fc_fd_master',con=engine,parse_dates=["acc_open_date","maturity_date"])
df_fd_master
```

```python
# Read fc_pre_facts
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_facts")
df_pre_facts=pd.read_sql('fc_pre_facts',con=engine)
df_pre_facts=df_pre_facts[["account_number","customer_code","in_lien_fd"]]
df_pre_facts
```

```python
df_fd_master_in_lien=df_fd_master[(df_fd_master["is_active"]==1) & (df_fd_master["maturity_date"]>run_date)]
df_fd_master_in_lien=df_fd_master_in_lien[["customer_code","is_in_lien"]]

# Change datatype of a field
df_fd_master_in_lien["is_in_lien"]=df_fd_master_in_lien["is_in_lien"].astype(int)
# print(df_fd_master_in_lien.dtypes)


df_fd_master_in_lien=df_fd_master_in_lien.groupby(["customer_code"]).agg(
    in_lien_fd=("is_in_lien","sum")
)
df_fd_master_in_lien.reset_index(inplace=True)

df_fd_master_in_lien=df_fd_master_in_lien[df_fd_master_in_lien["in_lien_fd"]!=0]
df_fd_master_in_lien
```

```python
df_qa_vs_dev=pd.merge(df_fd_master_in_lien,df_pre_facts,how="inner",on="customer_code",suffixes=('_qa','_dev'))
df_qa_vs_dev["status"]=np.select(
    [
        df_qa_vs_dev["in_lien_fd_qa"]==df_qa_vs_dev["in_lien_fd_dev"]
    ],
    [
        "pass"
    ],
    default="fail"
)
df_qa_vs_dev
```
