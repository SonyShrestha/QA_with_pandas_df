# Validation of days_till_maturity

```python
import sqlalchemy
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
run_date=datetime(2021,9,13)
run_date_plus_one_month=run_date+relativedelta(days=30)
run_date_plus_one_month
```

```python
# Read fc_account_master
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_account_master=pd.read_sql('fc_account_master',con=engine)
```

```python
# Get selected accounts from account master
df_account_master=df_account_master[df_account_master["active_flag"]==1]
```

```python
# Read fc_fd_master
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_fd_master=pd.read_sql('fc_fd_master',con=engine)
df_fd_master
```

```python
# Read dax_facts
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_facts")
df_pre_facts=pd.read_sql('fc_pre_facts',con=engine)
df_pre_facts
```

```python
# Get selected accounts from df_fd_master
df_fd_master["is_in_lien"]=df_fd_master["is_in_lien"].astype(int)
df_fd_master_selected=df_fd_master[(df_fd_master["is_active"]==1)&(df_fd_master["maturity_date"]>run_date) &(df_fd_master["is_in_lien"]==0)]
df_fd_master_selected=pd.merge(df_fd_master_selected,df_account_master,left_on="fd_nominee_acc",right_on="account_number",indicator=True,suffixes=('','_'))
df_fd_master_selected=df_fd_master_selected[df_fd_master_selected["_merge"]=='both']
df_fd_master_selected
```

```python
# Categorize records with maturity_days>=30 days and maturity_days<30 days
df_fd_master_gte_30=df_fd_master_selected[df_fd_master_selected["maturity_date"]>=run_date_plus_one_month]
df_fd_master_lt_30=df_fd_master_selected[df_fd_master_selected["maturity_date"]<run_date_plus_one_month]
```

```python
# sort records
df_fd_master_gte_30=df_fd_master_gte_30.sort_values(by=["customer_code","maturity_date"],ascending=False)
df_fd_master_lt_30=df_fd_master_lt_30.sort_values(by=["customer_code","maturity_date"],ascending=False)
```

```python
# aggregrate and get first value
df_fd_master_gte_30=df_fd_master_gte_30.groupby(["customer_code"]).agg(
        maturity_date=("maturity_date","first")
)
df_fd_master_lt_30=df_fd_master_lt_30.groupby(["customer_code"]).agg(
        maturity_date=("maturity_date","first")
)
df_fd_master_gte_30.reset_index(inplace=True)
df_fd_master_lt_30.reset_index(inplace=True)
```

```python
# full outer join
df_gte_lt_merged=pd.merge(df_fd_master_gte_30,df_fd_master_lt_30,how="outer",on="customer_code")
df_gte_lt_merged["maturity_date"]=np.select(
    [
        df_gte_lt_merged["maturity_date_x"].notna()
    ],
    [
        df_gte_lt_merged["maturity_date_x"]
    ],
    default= df_gte_lt_merged["maturity_date_y"]
)
df_gte_lt_merged["days_till_maturity"]=(df_gte_lt_merged["maturity_date"]-run_date).dt.days
df_gte_lt_merged
```

```python
# qa vs dev
df_qa_vs_dev=pd.merge(df_gte_lt_merged,df_pre_facts,how="inner",on="customer_code",suffixes=('_qa','_dev'))

df_qa_vs_dev=df_qa_vs_dev[["customer_code","days_till_maturity_qa","days_till_maturity_dev"]]

df_qa_vs_dev['status'] = np.select(
    [
        df_qa_vs_dev["days_till_maturity_qa"]==df_qa_vs_dev["days_till_maturity_dev"]
    ], 
    [
        'Pass'
    ], 
    default='Fail'
)

# Count of Distinct status
df_qa_vs_dev.value_counts("status")
```
