# Validation of cc_status

```python
import sqlalchemy
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.relativedelta import *
```

```python
end_date=datetime(2021,6,30)
end_date_sub_12_month=end_date-relativedelta(months=12)
end_date_sub_12_month
```

```python
# Read cc_due_report
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_cc_due_report=pd.read_sql('cc_due_report',con=engine,parse_dates=["bill_report_date"])
df_cc_due_report=df_cc_due_report[df_cc_due_report["bill_report_date"]>=end_date_sub_12_month]
df_cc_due_report["embossed_name"]=df_cc_due_report["embossed_name"].str.strip()
df_cc_due_report["mobile_number"]=df_cc_due_report["mobile_number"].str.strip()
df_cc_due_report
```

```python
# Read fc_account_master
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_account_master=pd.read_sql('fc_account_master',con=engine)
```

```python
# Read fc_pre_facts
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_facts")
df_pre_facts=pd.read_sql('fc_pre_facts',con=engine)
df_pre_facts=df_pre_facts[["account_number","customer_code","cc_status"]]
```

```python
merge_by_acc_num=pd.merge(df_account_master,df_cc_due_report,left_on="account_number",right_on="auto_debit_account",how="inner",suffixes=('','_'))
merge_by_acc_num=merge_by_acc_num[["account_number","customer_code","unpaid_status"]]
```

```python
merge_by_name_mob=pd.merge(df_account_master,df_cc_due_report,left_on=["account_name","mobile_number"],right_on=["embossed_name","mobile_number"],how="inner",suffixes=('','_'))
merge_by_name_mob=merge_by_name_mob[["account_number","customer_code","unpaid_status"]]
merge_by_name_mob
```

```python
union=pd.concat([merge_by_acc_num,merge_by_name_mob])
union=union.groupby(["account_number"]).agg(
    cc_status=("unpaid_status","max")
)
union.reset_index(inplace=True)
union
```

```python
# qa vs dev
df_qa_vs_dev=pd.merge(union,df_pre_facts,how="inner",on="account_number",suffixes=('_qa','_dev'))

df_qa_vs_dev=df_qa_vs_dev[["account_number","cc_status_qa","cc_status_dev"]]

df_qa_vs_dev['status'] = np.select(
    [
        df_qa_vs_dev["cc_status_qa"]==df_qa_vs_dev["cc_status_dev"]
    ], 
    [
        'Pass'
    ], 
    default='Fail'
)

# Count of Distinct status
df_qa_vs_dev.value_counts("status")
```
