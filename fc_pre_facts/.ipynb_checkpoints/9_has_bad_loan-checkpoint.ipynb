# Validation of has_bad_loan

```python
import sqlalchemy
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
bad_loan_provision=[2,3,4,5]
```

```python
# Read loan_master
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_loan_master=pd.read_sql('fc_loan_facts',con=engine)
df_loan_master
```

```python
# Read fc_pre_facts
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_facts")
df_pre_facts=pd.read_sql('fc_pre_facts',con=engine)
df_pre_facts=df_pre_facts[["account_number","customer_code","has_bad_loan"]]
df_pre_facts
```

```python
# df_loan_master_has_loan=df_loan_master[df_loan_master["provision_type"] in bad_provision_type]
df_loan_master_bad_loan=df_loan_master[["customer_code","provision_type"]]
df_loan_master_bad_loan["is_bad"]=np.select(
    [
        df_loan_master_bad_loan["provision_type"].isin(bad_loan_provision)
    ],
    [
        1
    ],
    default=0
)
df_loan_master_bad_loan
```

```python
df_loan_master_bad_loan=df_loan_master_bad_loan.groupby(["customer_code"]).agg(
    has_bad_loan=("is_bad","max")
)
df_loan_master_bad_loan.reset_index(inplace=True)
df_loan_master_bad_loan=df_loan_master_bad_loan[df_loan_master_bad_loan["has_bad_loan"]==1]
```

```python
df_loan_master_bad_loan
```

```python
df_has_bad_loan=pd.merge(df_loan_master_bad_loan,df_pre_facts,how="inner",on="customer_code",suffixes=('_qa','_dev'))
df_has_bad_loan["status"]=np.select(
    [
        df_has_bad_loan["has_bad_loan_qa"]==df_has_bad_loan["has_bad_loan_dev"]
    ],
    [
        "pass"
    ],
    default="fail"
)
df_has_bad_loan
```
