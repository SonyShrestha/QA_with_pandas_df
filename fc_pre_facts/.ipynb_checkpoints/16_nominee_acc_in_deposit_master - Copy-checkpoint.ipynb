# Validation of nominee_acc_dep_master_wo_in_lien_fd

```python
import sqlalchemy
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
run_date=datetime(2021,9,13)
```

```python
# Read transaction_base
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_account_master=pd.read_sql('fc_account_master',con=engine)
df_account_master=df_account_master[df_account_master["active_flag"]==1]
df_account_master
```

```python
# Read fc_fd_master
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_fd_master=pd.read_sql('fc_fd_master',con=engine)
df_fd_master
```

```python
df_fd_master["is_in_lien"]=df_fd_master["is_in_lien"].astype(int)
df_fd_master["is_in_lien"]==0
df_fd_master_active=df_fd_master[(df_fd_master["is_active"]==1)&(df_fd_master["maturity_date"]>run_date) & (df_fd_master["is_in_lien"]==0)]
df_fd_master_active
```

```python
df_fd_master["is_in_lien"]
```

```python
# Read fc_pre_facts
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_facts")
df_pre_facts=pd.read_sql('fc_pre_facts',con=engine)
df_pre_facts
```

```python
df_fd_deposit_merged=pd.merge(df_fd_master_active,df_account_master,left_on="fd_nominee_acc",right_on="account_number",how="inner")
df_fd_deposit_merged["nominee_acc_dep_master_wo_in_lien_fd"]=1
df_fd_deposit_merged=df_fd_deposit_merged.groupby(["customer_code_x"]).agg(
    nominee_acc_dep_master_wo_in_lien_fd=("nominee_acc_dep_master_wo_in_lien_fd","max")
)
df_fd_deposit_merged.reset_index(inplace=True)
df_fd_deposit_merged.rename(columns={"customer_code_x":"customer_code"},inplace=True)
df_fd_deposit_merged
```

```python
df_qa_vs_dev=pd.merge(df_fd_deposit_merged,df_pre_facts,how="inner",on="customer_code",suffixes=('_qa','_dev'))
df_qa_vs_dev=df_qa_vs_dev[["customer_code","nominee_acc_dep_master_wo_in_lien_fd_qa","nominee_acc_dep_master_wo_in_lien_fd_dev"]]
df_qa_vs_dev['status'] = np.select(
    [
        df_qa_vs_dev["nominee_acc_dep_master_wo_in_lien_fd_qa"]==df_qa_vs_dev["nominee_acc_dep_master_wo_in_lien_fd_dev"]
    ], 
    [
        'Pass'
    ], 
    default='Fail'
)

# Count of Distinct status
df_qa_vs_dev.value_counts("status")
```
