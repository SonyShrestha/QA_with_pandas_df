# Validation of has_active_fd

```python
import sqlalchemy
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
run_date=datetime(2021,9,10)
```

```python
# Read fd_master
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_fd_master=pd.read_sql('fc_fd_master',con=engine,parse_dates=["acc_open_date","maturity_date"])
df_fd_master
```

```python
# Read fc_pre_facts
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_facts")
df_pre_facts=pd.read_sql('fc_pre_facts',con=engine)
df_pre_facts=df_pre_facts[["account_number","customer_code","has_active_fd"]]
df_pre_facts
```

```python
df_fd_master_active_fd=df_fd_master[(df_fd_master["is_active"]==1) & (df_fd_master["maturity_date"]>run_date)]
df_fd_master_active_fd=df_fd_master_active_fd[["customer_code"]]
df_fd_master_active_fd["has_active_fd"]=1
df_fd_master_active_fd=df_fd_master_active_fd.groupby(["customer_code"]).agg(
    has_active_fd=("has_active_fd","max")
)
df_fd_master_active_fd.reset_index(inplace=True)
df_fd_master_active_fd
```

```python
df_qa_vs_dev=pd.merge(df_fd_master_active_fd,df_pre_facts,how="inner",on="customer_code",suffixes=('_qa','_dev'))
df_qa_vs_dev["status"]=np.select(
    [
        df_qa_vs_dev["has_active_fd_qa"]==df_qa_vs_dev["has_active_fd_dev"]
    ],
    [
        "pass"
    ],
    default="fail"
)
df_qa_vs_dev
```
