# Validation of recency 

```python
# Read table from mysql 
import sqlalchemy 
import pandas as pd
import numpy as np
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_raw")
df_account=pd.read_sql('fc_account_master',con=engine)
df_transaction=pd.read_sql('fc_transaction_base',con=engine)
```

```python
# Get selected columns of account master
df_account=df_account[["account_number","customer_code","account_name","mobile_number","product","product_category","acc_open_date","acc_closed_date","active_flag"]]
```

```python
# Get selected accounts of transaction base
df_transaction=df_transaction[["account_number","tran_date","lcy_amount","is_salary"]]
```

```python
df_transaction
```

```python
# Define end date
from datetime import datetime
end_date=datetime(2021,7,1)
```

```python
# Filter transactions with is_salary!=0
df_transaction_sal=df_transaction[df_transaction["is_salary"]!=0]
df_transaction_sal
```

```python
# Get max_salary transaction date for each account
df_transaction_max_sal_tran=df_transaction_sal.groupby(["account_number"]).agg(
    max_tran_date=("tran_date","max")
)
```

```python
# Date format
df_transaction_max_sal_tran["max_tran_date"]=pd.to_datetime(df_transaction_max_sal_tran["max_tran_date"].dt.strftime('%Y-%m-01'))

# Assign end date
df_transaction_max_sal_tran["end_date"]=end_date

# Get difference between end date and max_tran_date
df_transaction_max_sal_tran["month_diff"]=(df_transaction_max_sal_tran["end_date"].dt.year - df_transaction_max_sal_tran["max_tran_date"].dt.year) * 12 + (df_transaction_max_sal_tran["end_date"].dt.month  - df_transaction_max_sal_tran["max_tran_date"].dt.month )

# Find recency
df_transaction_max_sal_tran["recency"]=df_transaction_max_sal_tran["month_diff"]+1
df_transaction_max_sal_tran
```

```python
# Get records of fc_pre_facts as a dataframe
engine=sqlalchemy.create_engine("mysql+pymysql://sony:Extenso@123@10.13.189.55/dax_facts")
df_pre_facts=pd.read_sql('fc_pre_facts',con=engine)
```

```python
# Get selected columns of pre_facts
df_pre_facts_recency=df_pre_facts[["account_number","customer_code","recency"]]
df_pre_facts_recency
```

```python
# merge development and qa 
df_qa_vs_dev=pd.merge(df_pre_facts_recency,df_transaction_max_sal_tran,on='account_number',suffixes=('_qa','_dev'))

# Get selected columns
df_qa_vs_dev=df_qa_vs_dev[["account_number","customer_code","recency_qa","recency_dev"]]

# Status='pass' where dev_recency=qa_recency else Status="fail"
df_qa_vs_dev['status'] = np.select(
    [
        df_qa_vs_dev["recency_qa"]==df_qa_vs_dev["recency_dev"]
    ], 
    [
        'Pass'
    ], 
    default='Fail'
)

# Count of Distinct status
df_qa_vs_dev.value_counts("status")
```
